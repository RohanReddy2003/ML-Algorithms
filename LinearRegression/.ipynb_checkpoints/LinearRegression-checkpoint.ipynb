```python
df=pd.read_csv('HeightWeight.csv')
```

```python
# Normalizing the df
from sklearn.preprocessing import StandardScaler
x=df['Height'].values.reshape(-1,1)
y=df['Weight'].values.reshape(-1,1)
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
y
```

```python
# Calculating initial cost value
x=np.c_[(np.ones((len(x_scaled),1))),x_scaled]
```

```python
x.shape
```

```python
theta=np.zeros((x.shape[1],1))
```

```python
theta.shape
```

```python
def predict(x,theta):
    return x.dot(theta)

def compute_cost(x,y,theta):
    m=len(x)
    predictions=predict(x,theta)
    error=predictions-y
    cost=(1/(2*m))*np.sum(error**2)
    return cost

initial_cost=compute_cost(x,y,theta)
print(f'Initial cost: {initial_cost}')
```

```python
def gradient_descent(X, y, theta, alpha, num_iterations):
    m = len(y)
    cost_history = []

    for i in range(num_iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradient = (1 / m) * X.T.dot(errors)
        theta -= alpha * gradient

        cost = compute_cost(X, y, theta)
        cost_history.append(cost)
        if i % 100 == 0:
            print(f"Iteration {i}: Cost {cost}")

    return theta, cost_history

# Run gradient descent
theta, cost_history = gradient_descent(x, y, theta, 0.01, 1000)

print("Optimized theta:", theta)
print("Cost history:", cost_history)
```

```python
theta.shape
```

```python
plt.scatter(x_scaled,y,color='blue',label='Data points')
plt.plot(x_scaled,x.dot(theta),color='red',label='Best fit line')
plt.xlabel('Normalized Height')
plt.ylabel('Weights')
plt.title('Height vs. Weight with Best Fit Line')
plt.legend()
plt.show()
```
